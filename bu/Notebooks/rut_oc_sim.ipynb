```python
import sys
sys.path.append('/home/cbenitez')
import MTmp as mt
import matplotlib.pyplot as plt
import numpy as np
import oort_cons as oc
```

```python
marbles_dir = mt.tardir+'canicas/'
marbles_file = 'mp_marbles_06102020-045445'
```

```python
mar,sun_pos = mt.read_dat(marbles_dir, marbles_file)
```

```python
mar['gen'] = mar['creation_time']/4.8e13
```

```python
mar['gen'] = [int(mar['gen'].iloc[i])-2 for i in range(len(mar))]
```

```python
filter1 = (mar['output']==11) & (mar['gen']==1) & (abs(mar['z'])<=.1) & (abs(mar['z'])>=.05)
#filter1 = (mar['output']==11) & (mar['gen']==1) & (abs(mar['z'])<=.05) & (abs(mar['z'])>=.02)
mar_filt1 = mar[filter1]
mt.scatter(mar_filt1['x'],mar_filt1['y'],fig=None,figsize=(5,5),title='',label='',xlab='',ylab='',zlab='',xlims=(4,10),ylims=(-3,3),color='k',size=2)
```

```python
rs_test = np.array([7.45,-.28,0])
binx = int((max(mar_filt1['x'])-min(mar_filt1['x']))/.03)
biny = int((max(mar_filt1['y'])-min(mar_filt1['y']))/.03)
plt.figure(figsize=(5,5))
plt.hist2d(mar_filt1['x'],mar_filt1['y'],bins =[binx-1,biny-1])
plt.colorbar()
#plt.scatter([mass_center(mar_filt1)[0]],[mass_center(mar_filt1)[1]],c="orange")
plt.scatter([rs_test[0]],[rs_test[1]],c="orange")
plt.plot([0,rs_test[0]],[0,rs_test[1]],c="orange")
plt.show()
```
