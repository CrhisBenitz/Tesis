```python
def dens_colum(dd,zlim_index,z_central = None):
    if z_central ==None:
        z_central = dd.shape[0]/2
    zz = range(z_central-zlim_index,z_central+zlim_index+1)
    return sum(dd[z,:,:] for z in zz)/len(zz)
```

```python
mt.imshow(dens_colum(op10.dd[:,50:150:,100:200:],zlim_index=3),extentto=(op10.xx[100],op10.xx[200],op10.yy[50],op10.yy[150]),figsize=(8,8),logscale=True,colorbar=True,xlab =r"$x$ [kpc]",ylab=r"$y$ [kpc]",zlab = r"$n$ [$\mathrm{H_{cc}}$]")
circle = plt.Circle((8, 0), al/1000.,fill=False,color="r",alpha=1,lw=2)
plt.gca().add_artist(circle)
```
