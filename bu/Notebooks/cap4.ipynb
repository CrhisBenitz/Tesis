```python
def rho_lbd(r,l,b,rop,xv_LSR):
    r_LSR = xv_LSR[:3:]
    x,y,z = oc.cgalac2ccart([r,l,b],r_LSR)    
    rr = np.sqrt(x**2+y**2+z**2)
    nx,ny,nz = rop.nx,rop.ny,rop.nz
    x_max,y_max,z_max=rop.x_max,rop.y_max,rop.z_max
    x_min,y_min,z_min=rop.x_min,rop.y_min,rop.z_min
    
    cx_inf = (np.floor((nx-1)*(x - x_min)/(x_max-x_min))).astype(int)
    cx_sup = (np.ceil((nx-1)*(x - x_min)/(x_max-x_min))).astype(int)
    cy_inf = (np.floor((ny-1)*(y - y_min)/(y_max-y_min))).astype(int)
    cy_sup = (np.ceil((ny-1)*(y - y_min)/(y_max-y_min))).astype(int)
    cz_inf = (np.floor((nz-1)*(z - z_min)/(z_max-z_min))).astype(int)
    cz_sup = (np.ceil((nz-1)*(z - z_min)/(z_max-z_min))).astype(int)
    
    wx_sup = (x-rop.xx[cx_inf])/0.030000005651420025
    wx_inf = (rop.xx[cx_sup]-x)/0.030000005651420025    
    wy_sup = (y-rop.yy[cy_inf])/0.030000005651420025
    wy_inf = (rop.yy[cy_sup]-y)/0.030000005651420025   
    wz_sup = (z-rop.zz[cz_inf])/0.030000005651420025
    wz_inf = (rop.zz[cz_sup]-z)/0.030000005651420025
    
    dd = rop.dd
    densidad = dd[cz_inf,cy_inf,cx_inf]*wx_inf*wy_inf*wz_inf + dd[cz_sup,cy_inf,cx_inf]*wx_inf*wy_inf*wz_sup + dd[cz_inf,cy_sup,cx_inf]*wx_inf*wy_sup*wz_inf + dd[cz_inf,cy_inf,cx_sup]*wx_sup*wy_inf*wz_inf + dd[cz_sup,cy_sup,cx_inf]*wx_inf*wy_sup*wz_sup + dd[cz_sup,cy_inf,cx_sup]*wx_sup*wy_inf*wz_sup + dd[cz_inf,cy_sup,cx_sup]*wx_sup*wy_sup*wz_inf + dd[cz_sup,cy_sup,cx_sup]*wx_sup*wy_sup*wz_sup

    return densidad
```

```python
def rho_lbz(r,l,z,rop,xv_LSR):
    r_LSR = xv_LSR[:3:]
    b = np.arcsin(z/r)
    x,y,zf = oc.cgalac2ccart([r,l,b],r_LSR)    
    rr = np.sqrt(x**2+y**2+z**2)
    nx,ny,nz = rop.nx,rop.ny,rop.nz
    x_max,y_max,z_max=rop.x_max,rop.y_max,rop.z_max
    x_min,y_min,z_min=rop.x_min,rop.y_min,rop.z_min
    
    cx_inf = (np.floor((nx-1)*(x - x_min)/(x_max-x_min))).astype(int)
    cx_sup = (np.ceil((nx-1)*(x - x_min)/(x_max-x_min))).astype(int)
    cy_inf = (np.floor((ny-1)*(y - y_min)/(y_max-y_min))).astype(int)
    cy_sup = (np.ceil((ny-1)*(y - y_min)/(y_max-y_min))).astype(int)
    cz_inf = (np.floor((nz-1)*(z - z_min)/(z_max-z_min))).astype(int)
    cz_sup = (np.ceil((nz-1)*(z - z_min)/(z_max-z_min))).astype(int)
    
    wx_sup = (x-rop.xx[cx_inf])/0.030000005651420025
    wx_inf = (rop.xx[cx_sup]-x)/0.030000005651420025    
    wy_sup = (y-rop.yy[cy_inf])/0.030000005651420025
    wy_inf = (rop.yy[cy_sup]-y)/0.030000005651420025   
    wz_sup = (z-rop.zz[cz_inf])/0.030000005651420025
    wz_inf = (rop.zz[cz_sup]-z)/0.030000005651420025
    
    dd = rop.dd
    densidad = dd[cz_inf,cy_inf,cx_inf]*wx_inf*wy_inf*wz_inf + dd[cz_sup,cy_inf,cx_inf]*wx_inf*wy_inf*wz_sup + dd[cz_inf,cy_sup,cx_inf]*wx_inf*wy_sup*wz_inf + dd[cz_inf,cy_inf,cx_sup]*wx_sup*wy_inf*wz_inf + dd[cz_sup,cy_sup,cx_inf]*wx_inf*wy_sup*wz_sup + dd[cz_sup,cy_inf,cx_sup]*wx_sup*wy_inf*wz_sup + dd[cz_inf,cy_sup,cx_sup]*wx_sup*wy_sup*wz_inf + dd[cz_sup,cy_sup,cx_sup]*wx_sup*wy_sup*wz_sup

    return densidad
```
