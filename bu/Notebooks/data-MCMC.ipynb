```python
ll = np.arctan2(rr_uni_gal[:,1],rr_uni_gal[:,0])
bb = np.arctan2(rr_uni_gal[:,2],np.sqrt(rr_uni_gal[:,0]**2+rr_uni_gal[:,1]**2))
rr = np.array(1./df.parallax)
```

```python
df["l"] = ll
df["b"] = bb
```

```python
Picrs = np.array([-np.sin(df['ra']*np.pi/180),np.cos(df['ra']*np.pi/180),np.zeros_like(df['ra'])]).T
Qicrs = np.array([-np.cos(df['ra']*np.pi/180)*np.sin(df['dec']*np.pi/180),-np.sin(df['ra']*np.pi/180)*np.sin(df['dec']*np.pi/180),np.cos(df['dec']*np.pi/180)]).T
Pgal = np.array([-np.sin(df['l']*np.pi/180),np.cos(df['l']*np.pi/180),np.zeros_like(df['l'])]).T
Qgal = np.array([-np.cos(df['l']*np.pi/180)*np.sin(df['b']*np.pi/180),-np.sin(df['l']*np.pi/180)*np.sin(df['b']*np.pi/180),np.cos(df['b']*np.pi/180)]).T
mmicrs = np.array(Picrs*df['pmra'][:,np.newaxis] + Qicrs*df['pmdec'][:,np.newaxis])
mmgal = np.array([Ag.dot(m) for m in mmicrs])
mml = (Pgal*mmgal).sum(1)
mmb = (Qgal*mmgal).sum(1)
```
